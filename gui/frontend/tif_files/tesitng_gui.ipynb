```python
import rasterio
import numpy as np

with rasterio.open('bow_river_predictions_final_RF.tif') as src:
    rf = src.read(1)
with rasterio.open('bow_river_predictions_final_CNN.tif') as src:
    cnn = src.read(1)

print("Pixels that differ:", np.sum(rf != cnn))
print("Total pixels:", rf.size)
print("Percent different:", f"{(np.sum(rf != cnn) / rf.size * 100):.2f}%")
print("\nRF class distribution:", {v: int(np.sum(rf == v)) for v in np.unique(rf)})
print("CNN class distribution:", {v: int(np.sum(cnn == v)) for v in np.unique(cnn)})
```
